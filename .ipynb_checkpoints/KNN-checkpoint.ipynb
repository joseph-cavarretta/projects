```python
import pandas as pd
import numpy as np
import sklearn.neighbors
```

```python
sb_df = pd.read_csv('starbucks_us.csv')
```

```python
sb_df.head()
```

```python
sb_df = sb_df[['starbucksid', 'name', 'street1', 'city', 'countrysubdivisioncode', 'postalcode', 'lat', 'lon']]

sb_df.rename(columns={'starbucksid': 'starbucks_id', 'name': 'store_name', 'street1': 'store_address', 
                      'city': 'store_city', 'countrysubdivisioncode': 'store_state','postalcode': 'store_zip', 
                      'lat': 'store_lat', 'lon': 'store_long'}, inplace=True)
```

```python
sb_df.head()
```

```python
h_df = pd.read_csv("multifamily_physical_inspection_scores_0321.csv")
```

```python
h_df = h_df[['PROPERTY_ID', 'PROPERTY_NAME', 'ADDRESS', 'CITY', 'STATE_NAME', 'ZIPCODE', 'LATITUDE', 'LONGITUDE']]

h_df.columns = [col.lower() for col in h_df.columns]
```

```python
h_df.head()
```

```python
# add columns to convert latitude and longitude to radians
sb_df[['sb_lat_radians','sb_long_radians']] = (np.radians(sb_df.loc[:,['store_lat','store_long']]))

h_df[['lat_radians','long_radians']] = (np.radians(h_df.loc[:,['latitude','longitude']]))
```

```python
h_df.head()
```

```python
dist = sklearn.neighbors.DistanceMetric.get_metric('haversine')

# 3959 is the radius of the earth in miles, multiply to convert radians to miles
dist_matrix = (dist.pairwise(h_df[['lat_radians','long_radians']],
                             sb_df[['sb_lat_radians','sb_long_radians']])*3959)

# resultant df is a matrix of distances between all properties and starbucks stores
df_dist_matrix = (pd.DataFrame(dist_matrix, 
                               index=h_df['property_id'], 
                               columns=sb_df['starbucks_id']))
```

```python
# Unpivot this dataframe from wide format to long format.
# When you unpivot, the data in the pivot table becomes a
# column named 'value'. Rename this column to 'miles' for clarity.

df_dist_final = (pd.melt(df_dist_matrix.reset_index(), id_vars='property_id'))

df_dist_final.rename(columns={'value':'miles'}, inplace=True)
```

```python
df_dist_final.sort_values(['property_id', 'miles'], ascending = [False, True], inplace=True)
```

```python
df_dist_final.drop_duplicates(subset = ['property_id'], keep = 'first', inplace=True)
```

```python
df_dist_final.reset_index(drop = True, inplace=True)
```

```python
df_dist_final.head()
```

```python
# Avg distance to a starbucks nationwide
df_dist_final['miles'].mean()
```

```python
# Merge distances back into housing df
merged_df = h_df.merge(df_dist_final, on='property_id')
```

```python
merged_df.rename(columns={'miles': 'closest_starbucks (mi)'}, inplace=True)
```
